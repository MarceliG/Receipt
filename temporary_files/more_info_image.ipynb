```python
import cv2
import os
from config.definitions import PHOTOS_DIR
import matplotlib.pyplot as plt
import imutils
```

```python
image_name = "bill.jpg"
img = cv2.imread(os.path.realpath(os.path.join(PHOTOS_DIR, image_name)))
# width x height
image_size = f"Size of the original image: {img.shape[1]} x {img.shape[0]}"
image_ratio = f"Aspect ratio of original image: {img.shape[1]/img.shape[0]}"
print(image_size)
print(image_ratio)
plt.imshow(img)
```

```python
# rotate the image -- +90, 180, -90
from cv2 import ROTATE_90_CLOCKWISE, ROTATE_180, ROTATE_90_COUNTERCLOCKWISE

rotated_img = cv2.rotate(img, ROTATE_90_CLOCKWISE)
plt.title("Rotated")
plt.imshow(rotated_img)
```

```python
# rotate the image by different angle
# center: The central point of rotation in the given image
# angle: The angle of rotation. (Positive value denotes clockwise rotation while negative values denote anti-clockwise rotation)
# scale: Isotropic scaling factor

cols = img.shape[1]
rows = img.shape[0]
image_ratio = img.shape[1] / img.shape[0]

start_point = (0, 0)
end_point = (cols, rows)

color = (0, 255, 0)
thickness = 100

line = cv2.line(
    img=img, pt1=start_point, pt2=end_point, color=color, thickness=thickness
)

matrix = cv2.getRotationMatrix2D(
    center=(cols / 2, rows / 2), angle=-30, scale=image_ratio
)

rotate_30 = cv2.warpAffine(img, matrix, (cols, rows))


plt.title("Rotated")
plt.imshow(rotate_30)
```
