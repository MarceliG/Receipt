```python
from operation import load_image
from skimage.transform import radon
from process_image import remove_shadow, rotate, get_dataFrame_text_from_image
import matplotlib.pyplot as plt
from PIL import Image
from numpy import asarray, mean, array, blackman, argmax
from numpy.fft import rfft
import cv2
import imutils
import pytesseract
import pandas as pd
```

```python
image_name = "bill.jpg"
```

```python
# load image
loaded_image = load_image(image_name)
plt.imshow(loaded_image, cmap=plt.cm.gray)
```

```python
# remove shadow
removed_shadow = remove_shadow(loaded_image)
plt.imshow(removed_shadow, cmap=plt.cm.gray)
```

```python
# Do the radon transform and display the result
sinogram = radon(removed_shadow)
plt.imshow(sinogram.T, cmap=plt.cm.gray, aspect="auto")
```

```python
rotated_image = rotate(removed_shadow)

plt.title("Rotated")
plt.xticks([])
plt.yticks([])
plt.imshow(rotated_image, cmap=plt.cm.gray)
```

```python
text_data = get_dataFrame_text_from_image(rotated_image)
text_data
```
